# Multithreading

## Core idea

Single-threaded code:

Multithreaded idea:
- split 0..n-1 into chunks
- each thread handles one chunk
- wait for all threads to finish
- combine results if needed

So multithreading is about **task-level** or **data-level parallel execution across CPU cores**.

### Why it helps

Modern CPUs have multiple cores.

If your program uses only one thread, it may use only one core well.

With multithreading, the program can use several cores at once.

Typical use cases:
- array and matrix processing
- image processing
- simulations
- data pipelines
- file processing
- server handling many requests
- ML preprocessing or CPU inference kernels

**Mental model**
- Suppose you need to process 100 million elements.

**One thread**
- One worker does everything.

**Four threads**
- Each worker does about 25 million elements.

If the work is independent and the machine has enough cores, it can be much faster.

## Basic concepts

1. Thread creation

In C++, you can create a thread with std::thread.

2. Join

You must usually call join() so the main thread waits for worker threads to finish.

3. Shared memory

Threads in the same process share memory, which is powerful but dangerous.

4. Data race

If multiple threads read/write the same variable unsafely, the result may be wrong.

5. Synchronization

Use tools like:
- ```std::mutex```
- ```std::lock_guard```
- ```std::atomic```

to protect shared data.
