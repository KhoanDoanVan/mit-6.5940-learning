# Loop Tiling

## Core idea

Suppose you have a 2D loop:

Here:
- ```B``` = tile size
- ```ii, jj``` = which tile you are in
- ```i, j``` = positions inside the tile

So instead of processing the whole big matrix row by row, you process it block by block.

### Why it helps

Modern CPUs are fast, but RAM is slow.


Cache is small, but very fast.

**If the data you need fits in cache, performance improves a lot.**

Loop tiling helps because:
- it keeps a small working set in cache
- it increases data reuse
- it reduces cache misses
- it often improves bandwidth efficiency

So the optimization is really about **locality**.

### Intuition

Imagine a huge matrix.

Without tiling:
- you touch data
- move far away
- come back later
- by then, the old data is no longer in cache

With tiling:
- you load a small block
- do as much work as possible on that block
- then move to the next block

That is the win.
