## Core concept

Modern CPUs are much faster than main memory, so performance often depends on how data is laid out in memory and how you access it.

In C and C++, a 2D array is typically stored in **row-major order**. That means elements of the same row are next to each other in memory.

If your inner loop walks across columns of the same row, memory access is contiguous, which is usually fast.

If your inner loop jumps between rows, memory access can become strided, which is often slower.
