# Loop Unrolling

## Core idea

A normal loop:

does this every iteration:
- compare ```i < N```
- branch back to loop start
- increment ```i```
- do one useful operation

With unrolling by 4:

Now one loop iteration does 4 useful operations, so:
- fewer branches
- fewer index updates
- less loop overhead
- often more chance for compiler optimization and SIMD

**So loop unrolling is not about changing the algorithm.**

**It is about reducing the overhead around the algorithm.**

### Intuition

Suppose the original loop runs 1000 times.

If you unroll by 4, the loop body runs only about 250 times.

That means the program performs:
- about 4x fewer branch checks
- about 4x fewer loop counter increments

The useful arithmetic work stays the same, but the control overhead drops.
