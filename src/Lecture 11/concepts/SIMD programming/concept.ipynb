# SIMD (Single Instruction, Multiple Data.)

## Core idea

Normal scaler code:

This conceptually does:
- ```c[0] = a[0] + b[0]```
- ```c[1] = a[1] + b[1]```
- ```c[2] = a[2] + b[2]```
- ...

one element at a time.

With SIMD, the CPU may process several elements together:
- 4 floats at once
- 8 floats at once
- 16 bytes at once
- more, depending on AVX, SSE, NEON, etc.

So the operation stays the same, but the throughput increases.

### Why it helps

SIMD is useful when:
- the same math is applied to many elements
- data is laid out regularly
- iterations are independent

Typical use cases:
- vector addition
- matrix operations
- image processing
- signal processing
- neural network kernels
- audio processing

### Simple intuition

Suppose you want to add two arrays:

Scalar execution does:
```1 + 10```
```2 + 20```
```3 + 30```
```4 + 40```
SIMD can load both vectors and do all four additions together.
That is the essence of SIMD.
