## Pooling

1. Định nghĩa

Pooling là phép toán downsampling không có tham số học, hoạt động bằng cách tổng hợp (aggregate) giá trị trong một vùng nhỏ.

Các loại phổ biến:
- Max Pooling → lấy giá trị lớn nhất
- Average Pooling → lấy trung bình
- Global Pooling → reduce toàn bộ feature map về 1×1

2. Ưu điểm

(1) Không có tham số
- Không cần train
- Không tốn memory weight

(2) Giảm overfitting
- Làm mất chi tiết nhỏ → tránh model overfit noise

(3) Translation invariance
- Object dịch nhẹ vẫn nhận ra được

3. Nhược điểm

(1) Mất thông tin không kiểm soát
- Không học được cái gì cần giữ
- Có thể mất feature quan trọng

(2) Mất spatial precision
- Rất nguy hiểm với:
    - detection
    - segmentation
    - AR

(3) Không adapt theo data
- Fixed rule → không tối ưu theo task
